```python
df=pd.read_csv('CP1_data.csv')
```

```python
import re
df['passage'] = [re.sub('[^A-Za-z0-9가-힣]', '', s) for s in df['passage']] 
df
```

```python
tokenized_texts=[]
for text in df['passage']:
    tokenized_text = tokenizer.encode(text,truncation=True, add_special_tokens=True)
    tokenized_texts.append(tokenized_text)
   
df['tokenized_text'] = tokenized_texts
```

```python
tokenized_sums=[]
for text in df['summary']:
    tokenized_summary = tokenizer.encode(text,truncation=True, add_special_tokens=True)
    tokenized_sums.append(tokenized_summary)
   
df['tokenized_sums'] = tokenized_sums

df
```
